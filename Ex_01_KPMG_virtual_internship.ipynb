#### 3.5 Chuẩn bị các cột dữ liệu cho model RFM 

##### 3.5.1 Lấy dữ liệu các giao dịch đã thực hiện Approved

```python
#Lấy dữ liệu các giao dịch đã được chấp nhận
Transactions = Transactions.loc[~(Trans.order_status == "Cancalled")]
Transactions
```

##### 3.5.2 Tạo cột các giao dịch recency cho mô hình RFM
- Lấy ngày giao dịch lớn nhất trong bộ dữ liệu trừ cho ngày giao dịch.

```python
# Lấy các giao dịch gần đây
recent_purchase = Transactions['transaction_date'].max()
Transactions['recency'] = recent_purchase - Transactions['transaction_date']

#Convert date
Transactions['recency'] /= np.timedelta64(1,'D')
Transactions.head()
```

##### 3.5.3 Gán nhãn phân loại cho recency => recency_category

```python
#Give a score for recency. 1 being the lowest and 4 the highest.
Transactions['recency_category'] = pd.qcut(Transactions['recency'], q=4, labels= [4,3,2,1])
```

```python
def customers(r,f,m):
    
    if (r == 4) and (f in (3, 4)) and (m in (3,4)):
        return 'Top'
    
    elif (r == 4) and (f == 2) and (m in (1,2,3,4)):
        return 'Active'
    
    elif (r == 4) and (f in (3,4)) and (m in (1,2)):
        return 'Active'
    
    elif (r == 3) and (f in (2,3,4)) and (m in (3,4)):
        return 'Emerging'
    
    elif (r in (3,4)) and (f == 1) and (m in (1,2,3,4)):
        return 'Unsteady'
    
    elif (r in (3,4)) and (f in (2,3,4)) and (m in (1,2)):
        return 'Unsteady'
        
    elif (r == 2) and (f in (1,2)) and (m in (1,2)):
        return 'Potential lost'
    
    elif (r == 1) and (f in (1,2)) and (m in (3,4)):
        return 'Inactive'
    
    elif (r == 1) and (f in (1,2)) and (m in (1,2)):
        return 'Lost'
    
    elif (r == 2) and (f in (1,2,3,4)) and (m in (3,4)):
        return 'At risk'
    
    elif (r == 2) and (f in (3,4)) and (m in (1,2)):
        return 'At risk'
        
    elif (r == 1) and (f in (3,4)) and (m in (1,2,3,4)):
        return'At risk'
    
```
